```python
import random
import csv

# Define the constraints
min_distance_to_first_row = 0.5
max_distance_to_last_row = 5
min_seats_per_row = 1
min_seats_per_cluster = 1
max_seats_per_cluster = 10  # Adjust this to an appropriate maximum value
total_number_of_seats_in_room = 0  # Initialize total number of seats in the room

# Create an empty list to store the dataset
dataset = []

# Define the number of data points you want in your dataset
num_data_points = 2000  # You can change this number as needed

for _ in range(num_data_points):
    # Randomly generate video wall dimensions
    height_of_video_wall = random.uniform(2, 5)  # Specify a plausible range
    width_of_video_wall = random.uniform(3, 8)  # Specify a plausible range

    # Randomly generate room dimensions larger than the video wall
    room_length = random.uniform(width_of_video_wall + 2, 2 * width_of_video_wall + 5)
    room_width = random.uniform(height_of_video_wall + 2, 2 * height_of_video_wall + 5)
    room_height = random.uniform(3, 5)  # Specify a plausible range

    # Randomly choose seating type (alternating between "row" and "cluster")
    seating_type = "row" if len(dataset) % 2 == 0 else "cluster"

    if seating_type == "row":
        # Calculate the maximum seats per row based on room length
        max_seats_per_row = int(room_length)

        # Randomly generate seats_per_row within the constraints
        seats_per_row = random.randint(min_seats_per_row, max_seats_per_row)

        # Calculate the number of rows based on the number of seats per row
        number_of_rows = max(1, int(room_length / seats_per_row))
        
        number_of_seats = seats_per_row * number_of_rows  # Calculate the total number of seats for row seating

        seats_per_cluster = 0  # Initialize seats per cluster (not used for rows)
        number_of_clusters = 0  # Initialize number of clusters (not used for rows)
    else:
        # Calculate the maximum seats per cluster based on the total number of seats in the room
        max_seats_per_cluster = max(min_seats_per_cluster, int(total_number_of_seats_in_room / 10))

        # Randomly generate seats_per_cluster within the constraints
        seats_per_cluster = random.randint(min_seats_per_cluster, max_seats_per_cluster)

        # Calculate the number of clusters based on the number of seats per cluster
        number_of_clusters = max(1, int(room_length / (seats_per_cluster + 1)))  # Ensure clusters fit in the room
        
        number_of_seats = seats_per_cluster * number_of_clusters  # Calculate the total number of seats for cluster seating

        seats_per_row = 0  # Initialize seats per row (not used for clusters)
        number_of_rows = 0  # Initialize number of rows (not used for clusters)

    # Randomly generate distance_from_video_wall within the constraints
    min_distance_multiplier = min_distance_to_first_row * height_of_video_wall
    max_distance_multiplier = max_distance_to_last_row * height_of_video_wall
    distance_from_video_wall = random.uniform(min_distance_multiplier, min(max_distance_multiplier, room_length - 0.01))

    # Append the data point to the dataset
    data_point = (height_of_video_wall, width_of_video_wall, room_length, room_width, room_height, seating_type, seats_per_row, number_of_rows, seats_per_cluster, number_of_clusters, distance_from_video_wall, number_of_seats)
    dataset.append(data_point)

# Define the CSV file name
csv_file_name = "video_wall_data.csv"

# Save the dataset to a CSV file
with open(csv_file_name, mode='w', newline='') as csv_file:
    fieldnames = ['height_of_video_wall', 'width_of_video_wall', 'room_length', 'room_width', 'room_height', 'seating_type', 'seats_per_row', 'number_of_rows', 'seats_per_cluster', 'number_of_clusters', 'distance_from_video_wall', 'number_of_seats']
    writer = csv.writer(csv_file)
    
    # Write the header row
    writer.writerow(fieldnames)
    
    # Write the data rows
    for data_point in dataset:
        writer.writerow(data_point)

print(f'Data saved to {csv_file_name}')
```
